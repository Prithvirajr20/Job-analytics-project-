### EDA

```python
#job category count
plt.figure(figsize = (16,10))
sns.countplot(x ='job_category', data = df, color='blue')
plt.xticks(rotation=90)
plt.show()
```

above chart shows that Data science and Research jobs are highly available in the job market

```python
#experience_level count
plt.figure(figsize=(16, 10))
sns.countplot(y='experience_level', data=df, color='grey')
plt.title('experience_level Counts')
plt.xlabel('Count')
plt.ylabel('experience_level Type')
plt.show()
```

Above chart represents that Senior Level jobs are high in numbers as compared to entry level, so its very hard to get a job as a fresher

```python

# Column chart for employment type counts
plt.figure(figsize=(16, 10))
sns.countplot(y='employment_type', data=df, color='skyblue')
plt.title('Employment Type Counts')
plt.xlabel('Count')
plt.ylabel('Employment Type')
plt.show()
```

above chart represents that Full Time jobs are high in numbers

```python
# Count the occurrences of each work setting
work_setting_counts = df['work_setting'].value_counts()

# Create a donut chart
plt.figure(figsize=(10, 8))
plt.pie(work_setting_counts, labels=work_setting_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4, edgecolor='w'))

# Draw a circle in the center to create a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Work Setting Distribution')
plt.show()
```

its shows that most of the jobs requires onsite work setting followed by remote mode of working

```python
#company_size          
plt.figure(figsize = (16,10))
sns.countplot(x ='company_size', data = df,color='skyblue')
plt.xticks(rotation=0)
plt.show()
```

1.Lets check the growth of job market over years

```python
# Count the number of jobs for each year
job_counts = df['work_year'].value_counts().sort_index()

# Plotting the data
plt.plot(job_counts.index, job_counts.values, marker='o', linestyle='-', color='b')
plt.title('Job Count Based on Year')
plt.xlabel('Work Year')
plt.ylabel('Job Count')
plt.show()
```

From the above line chart it shows that the jobs available in market are increasing over the period of years 

```python
# Count the number of jobs for each year and experience level
job_counts = df.groupby(['work_year', 'experience_level']).size().unstack().fillna(0)

# Plotting the data
job_counts.plot(kind='line', marker='o', linestyle='-', figsize=(10, 6))
plt.title('Job Count Based on Year and Experience Level')
plt.xlabel('Work Year')
plt.ylabel('Job Count')
plt.legend(title='Experience Level', loc='upper right')
plt.show()
```

Senior level roles are highly increasing over the years as compared to executive and entry level roles


```python
# Violin plot for salary range based on job category
plt.figure(figsize=(12, 8))
sns.violinplot(x='job_category', y='salary_in_usd', data=df)
plt.title('Salary Range by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

The salary range for all job categories is broad, with a spread of  $ 300000. 


This suggests that there is a lot of variation in salaries within each job category.


The highest paid job category is data engineering, with a median salary of $ 300000.


The lowest paid job category is data management and strategy, with a median salary of $ 100000.


There is a significant overlap in salary ranges between job categories. For example, the salary range for data science and research overlaps with the salary range for both data engineering and data analysis.

```python
# Violin plot for salary range based on experience level
plt.figure(figsize=(12, 8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary Range by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Group by job category and experience level, then calculate the average salary
grouped_data = df.groupby(['job_category', 'experience_level'])['salary_in_usd'].mean().reset_index()

# Bar plot for average salary based on job category and experience level
plt.figure(figsize=(14, 8))
sns.barplot(x='job_category', y='salary_in_usd', hue='experience_level', data=grouped_data, palette='Set1')
plt.title('Average Salary by Job Category and Experience Level')
plt.xlabel('Job Category')
plt.ylabel('Average Salary')
plt.xticks(rotation = 90)
plt.show()
```

The highest paid job category is Data Science and Research, with an average salary of $175,000 for senior-level workers. 

The lowest paid job category is Data Management and Strategy, with an average salary of $100,000 for senior-level workers.
Salaries increase with experience level for all job categories. 

For example, the average salary for entry-level data scientists is $75000 , while the average salary for senior-level data scientists is $175000.


There is a significant overlap in salary ranges between job categories. For example, the salary range for data analysts overlaps with the salary range for both data scientists and data engineers.

```python
# Find the maximum salary for each job category
max_salary_by_category = df.groupby('job_category')['salary_in_usd'].max().sort_values(ascending=False)

# Bar plot for top job categories based on the highest salary
plt.figure(figsize=(12, 8))
sns.barplot(x=max_salary_by_category.values, y=max_salary_by_category.index, palette='viridis')
plt.title('Top Job Categories Based on Highest Salary')
plt.xlabel('Highest Salary')
plt.ylabel('Job Category')
plt.show()
```

The highest paid job category is Data science and research, with a maximum salary of $400,000. This role is responsible for overseeing all of an organization's data assets and making strategic decisions about how to use data to improve the business.

Other high-paying job categories include Machine Learning Engineer, Data Scientist, and Data Architect. These roles require a high level of technical expertise and the ability to solve complex problems using data.

It is also important to note that salaries can vary depending on factors such as location, experience, and employer.

```python
# Find the maximum salary for each country
max_salary_by_country = df.groupby('employee_residence')['salary_in_usd'].max().sort_values(ascending=False)

# Select the top 5 countries
top_5_countries = max_salary_by_country.head(5)

# Bar plot for top 5 countries based on highest salary
plt.figure(figsize=(12, 8))
sns.barplot(x=top_5_countries.values, y=top_5_countries.index, palette='Set2')
plt.title('Top 5 Countries with Highest Salary')
plt.xlabel('Highest Salary')
plt.ylabel('Country')
plt.show()
```

Above are the top 5 countries based on the maximum salary and USA comes first

```python
# Find the maximum salary for each company size
max_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].max().sort_values(ascending=False)

# Bar plot for maximum salary based on company size
plt.figure(figsize=(12, 8))
sns.barplot(x=max_salary_by_company_size.index, y=max_salary_by_company_size.values, palette='viridis')
plt.title('Maximum Salary Based on Company Size')
plt.xlabel('Company Size')
plt.ylabel('Maximum Salary')
plt.show()
```

Above chart clearly states that the medium size companies are offering better salary than large and small sized companies

```python

# Find the maximum salary for each work setting
max_salary_by_work_setting = df.groupby('work_setting')['salary_in_usd'].max().sort_values(ascending=False)

# Pie chart for maximum salary based on work setting
plt.figure(figsize=(8, 8))
plt.pie(max_salary_by_work_setting, labels=max_salary_by_work_setting.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))

plt.title('Maximum Salary Based on Work Setting')
plt.show()
```

From the above pie chart we can say that onsite(in-person) working mode providing better salary that remote and hybrid
